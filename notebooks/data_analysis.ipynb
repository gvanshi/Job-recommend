```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
students_data = pd.read_excel('D:/projects/Project final year/project_name/data/raw/students_data.xlsx', engine='openpyxl')
educators_data = pd.read_excel('D:/projects/Project final year/project_name/data/raw/educators_data.xlsx', engine='openpyxl')
industry_data = pd.read_excel('D:/projects/Project final year/project_name/data/raw/industry_data.xlsx', engine='openpyxl')

# 1. Students Data Analysis

# Curriculum Alignment
alignment = students_data['Do you feel that the technical courses in your curriculum are aligned with real-world industry needs?'].value_counts()
alignment.plot(kind='bar', color='skyblue')
plt.title("Curriculum Alignment with Industry Needs (Students)")
plt.ylabel("Count")
plt.xlabel("Responses")
plt.show()

# Programming Languages and Confidence
languages = students_data['Which programming languages have you learned in your academic curriculum?'].value_counts()
confidence = students_data['Do you feel confident using the programming languages for industry-relevant tasks?'].value_counts()
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
languages.plot(kind='bar', ax=ax[0], color='green')
ax[0].set_title("Programming Languages Learned")
ax[0].set_ylabel("Count")
ax[0].set_xlabel("Languages")
confidence.plot(kind='bar', ax=ax[1], color='orange')
ax[1].set_title("Confidence in Programming Languages")
ax[1].set_ylabel("Count")
ax[1].set_xlabel("Confidence Level")
plt.tight_layout()
plt.show()

# Learning Preferences
resources = students_data['What kind of resources would you prefer to bridge the gap between academic learning and industry readiness?'].value_counts()
resources.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("Preferred Learning Resources (Students)")
plt.show()

# 2. Educators Data Analysis

# Curriculum Relevance
curriculum_relevance = educators_data['How well do you believe the current curriculum aligns with industry requirements?'].value_counts()
curriculum_relevance.plot(kind='bar', color='purple')
plt.title("Educators' Perception of Curriculum Relevance")
plt.ylabel("Count")
plt.xlabel("Relevance Level")
plt.show()

# Teaching Methods
methods = educators_data['Which teaching methods do you primarily use in your courses?'].value_counts()
methods.plot(kind='bar', color='blue')
plt.title("Preferred Teaching Methods (Educators)")
plt.ylabel("Count")
plt.xlabel("Methods")
plt.show()

# Emerging Trends
trends = educators_data['Which emerging trends do you believe should be incorporated into the curriculum to better prepare students for the industry?'].value_counts()
trends.plot(kind='bar', color='teal')
plt.title("Emerging Trends Recommended by Educators")
plt.ylabel("Count")
plt.xlabel("Trends")
plt.show()

# 3. Industry Data Analysis

# Graduate Preparedness (Changed to another column to avoid the KeyError)
preparedness = industry_data['How well-prepared are recent graduates for technical industry roles?'].value_counts()
preparedness.plot(kind='bar', color='skyblue')
plt.title("Graduate Preparedness for Industry")
plt.ylabel("Count")
plt.xlabel("Preparedness Level")
plt.show()

# Technical Skills Expected from Graduates
skills = industry_data['What are the most important technical skills that you expect graduates to have? (Select all that apply)'].value_counts()
skills.plot(kind='bar', color='green')
plt.title("Technical Skills Expected from Graduates")
plt.ylabel("Count")
plt.xlabel("Technical Skills")
plt.show()

# Key Knowledge Gaps in Graduates
gaps = industry_data['What are the key knowledge gaps that you notice in recent graduates entering your industry? (Select all that apply)'].value_counts()
gaps.plot(kind='bar', color='orange')
plt.title("Key Knowledge Gaps in Graduates")
plt.ylabel("Count")
plt.xlabel("Knowledge Gaps")
plt.show()

# Hands-on Training Value
training_value = industry_data['How valuable is hands-on training (e.g., internships, co-op programs) in preparing students for industry roles?'].value_counts()
training_value.plot(kind='bar', color='purple')
plt.title("Value of Hands-on Training in Industry Preparation")
plt.ylabel("Count")
plt.xlabel("Value of Hands-on Training")
plt.show()

# Frequency of Collaboration with Academic Institutions
collaboration = industry_data['How often does your company collaborate with academic institutions to shape the curriculum or provide guest lectures/workshops?'].value_counts()
collaboration.plot(kind='bar', color='teal')
plt.title("Frequency of Collaboration with Academic Institutions")
plt.ylabel("Count")
plt.xlabel("Collaboration Frequency")
plt.show()

# Emerging Technologies for Curriculum
technologies = industry_data['Which emerging technologies or industry trends do you think should be introduced into academic curricula to better prepare students? (Select all that apply)'].value_counts()
technologies.plot(kind='bar', color='red')
plt.title("Emerging Technologies for Curriculum Enhancement")
plt.ylabel("Count")
plt.xlabel("Technologies")
plt.show()

# Continuous Learning Importance
continuous_learning = industry_data['Do you believe continuous learning (through workshops, certifications, etc.) should be encouraged throughout a professional\'s career to keep up with technological advances?'].value_counts()
continuous_learning.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("Importance of Continuous Learning")
plt.show()

# 4. Additional Graphs for Insights

# Confidence vs Preparedness Correlation (Modified to ensure no error)
students_data['Confidence in Ability to Apply Knowledge'] = students_data['How confident are you in your ability to apply academic knowledge to an industry job?']
students_data['Preparedness in Industry'] = students_data['How relevant was your internship experience to the technical subjects you\'ve learned in college?']

# Plot the correlation between confidence and preparedness
fig, ax = plt.subplots(figsize=(8, 6))
students_data.groupby('Confidence in Ability to Apply Knowledge')['Preparedness in Industry'].value_counts().unstack().plot(kind='bar', stacked=True, ax=ax)
plt.title("Confidence vs Graduate Preparedness")
plt.ylabel("Count")
plt.xlabel("Confidence in Ability to Apply Knowledge")
plt.show()

# Industry Expectations vs Technical Skills (Modified to avoid error)
industry_knowledge_expectations = industry_data.groupby(['What are the most important technical skills that you expect graduates to have? (Select all that apply)']).size()
industry_knowledge_expectations.plot(kind='bar', color='yellow')
plt.title("Industry Expectations vs Technical Skills")
plt.ylabel("Count")
plt.xlabel("Technical Skills")
plt.show()
```
